## **List of Squares**
**Statement**: 
For a given integer N, print all the squares of integer numbers where the square is less than or equal to N, in ascending order.

```python
#1 Read an integer:
n = int(input())
#2 Set an integer:
x = 1
#3 Set a while loop condition so that the program can print the squared numbers until the condition is false:
while x**2 <= n:
#4 Print the squared numbers:
    print(x**2)
#5 Add 1 each time the loop happens so that the loop can end:
    x += 1
```

## **Least Divisor**
**Statement**: 
Given an integer not less than 2. Print its smallest integer divisor greater than 1.


```python
#1 Read an integer, set an base integer for the loop:
n = int(input())
x = 2

#2 Create a while condition to find the smallest divisor greater than 1, and add an 1 everytime the loop happens
while n % x != 0:
    x += 1
    
#3 Print
print(x)
```

## **The power of 2**
**Statement**: 
For a given integer N, find the greatest integer x where 2^x is less than or equal to N. Print the exponent value and the result of the expression 2^x.
Don't use the operation **.


```python
#1 Read and set two base inputs
n = int(input())
x = 2
y = 1

#2 Create a while condition for the greatest integer x and 2^x
while x <= n:
    x *= 2    
    y += 1

#3 Print
print(y - 1, x // 2)
```

## **Morning jog**
**Statement**: 
As a future athlete you just started your practice for an upcoming event.

 Given that on the first day you run x miles, and by the event you must be able to run y miles.
 
Calculate the number of days required for you to finally reach the required distance for the event, if you increases your distance each day by 10% from the previous day.

Print one integer representing the number of days to reach the required distance.


```python
#1 Read and set an integer
x = int(input())
y = int(input())
z = 1

#2 While loop condition for calculating the distance
while x < y:
    x *= 1.1
    z += 1

#3 Print the days to reach the distance
print(z)
```

## **The sum of the sequence**

## **The maximum of the sequence**
**Statement**: 
Determine the maximum of all elements in the sequence, ending with the number 0.


```python
#1 Read and set an integer
x = int(input())
y = 0

#2 Create a while loop condition to find the maximum of sequence 
while x != 0:
    if x > y:
        y = x
    x = int(input())
    
#3 Print
print(y)
```

## **The number of elements equal to the maximum**
**Statement**: 
A sequence consists of integer numbers and ends with the number 0. Determine how many elements of this sequence are equal to its largest element.


```python
#1 Read and set an integer
x = int(input())
y = 0 
z = 1

#2 Create while loop condition to see how many numbers are equal to the maximum of sequence
while x != 0:
    if x > y:
        y = x
        z = 1
    elif x == y:
        z += 1
    x = int(input())

#3 Print
print(z)
```
